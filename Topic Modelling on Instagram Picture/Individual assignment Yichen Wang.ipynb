```python
import pandas as pd

df = pd.read_excel("GV_Output.xlsx")
df
```

```python
#!pip install gensim
import gensim
from gensim.utils import simple_preprocess
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
import pyLDAvis.gensim_models
from pprint import pprint
import gensim.corpora as corpora
```

```python
label = df["Labels"].tolist()
label
```

```python
for i in range(len(label)):
    label[i] = label[i].split()
label
```

```python
# Create Dictionary
id2word = corpora.Dictionary(label)
# Create Corpus
texts = label
# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts]
# View
```

**LDA model training**

```python
# number of topics
num_topics = 4
# Build LDA model
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=id2word,
                                       num_topics=num_topics,
                                      random_state=35)
# Print the Keyword in the 10 topics
pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```

```python
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim_models.prepare(lda_model, corpus, dictionary=lda_model.id2word)
vis
```

```python
for topicid in range(num_topics):    
    print("for cluster",topicid,", the top 25 labels are:")
    print(lda_model.show_topic(topicid, topn=25))
    print("-------------------------------------------------------------------")
```

**Now sort the data from high to low engagement (# of comments), and take the highest and the lowest quartiles. What are the main differences in the average topic weights of images across the two quartiles (e.g., greater weight of some topics in the highest versus lowest engagement quartiles)? Show the main results in a table.**

```python
temp1 = pd.read_excel("GV_Output.xlsx")
temp2 = pd.read_excel("Insta_download.xlsx")
final = pd.merge(temp1, temp2, on="URL")
final = final.drop(columns = "Caption")
final.sort_values(by=['Comments'], inplace=True)
final = final.dropna()
final = final.reset_index()
final.drop(columns = "index",inplace=True)
final
```

```python
length = len(final)
length
```

**Highest quartile: top 25%
Lowest quartile: last 25%**

```python
high = final.iloc[int(length*0.75):]
low = final.iloc[:int(length*0.25)]
```

```python
high_label = high["Labels"].reset_index().drop(columns = "index")
low_label = low["Labels"].reset_index().drop(columns = "index")
```

```python
for i in range(len(high_label)):
    high_label["Labels"][i] = high_label["Labels"][i].split()

for i in range(len(low_label)):
    low_label["Labels"][i] = low_label["Labels"][i].split()
```

```python
high_label_list = high_label["Labels"].tolist()
low_label_list = low_label["Labels"].tolist()
```

```python
id2word_high = corpora.Dictionary(high_label_list)
# Create Corpus
texts_high = high_label_list
# Term Document Frequency
corpus_high = [id2word_high.doc2bow(text) for text in texts_high]

id2word_low = corpora.Dictionary(low_label_list)
# Create Corpus
texts_low = low_label_list
# Term Document Frequency
corpus_low = [id2word_low.doc2bow(text) for text in texts_low]
```

```python
result_high = lda_model.get_document_topics(corpus_high)
result_low = lda_model.get_document_topics(corpus_low)
```

```python
high_topics = lda_model[result_high]
low_topics = lda_model[result_low]

for topic in high_topics:
    print(topic)
```

```python
topic_1=0
topic_2=0
topic_3=0
topic_4=0
weight_high = [topic_1,topic_2,topic_3,topic_4]
weight_low = [topic_1,topic_2,topic_3,topic_4]
```

```python
for i in range(len(high_topics)):
    for j in range(len(weight_high)):
        weight_high[j] = weight_high[j]+ high_topics[i][j][1]

for i in range(len(low_topics)):
    for j in range(len(weight_low)):
        weight_low[j] = weight_low[j]+ low_topics[i][j][1]
```

```python
for i in range(len(weight_high)):
    weight_high[i] = weight_high[i]/len(high_topics)
print("topic weights for high quartile:",weight_high[0],weight_high[1],weight_high[2],weight_high[3])

for i in range(len(weight_low)):
    weight_low[i] = weight_low[i]/len(low_topics)
print("topic weights for low quartile:",weight_low[0],weight_low[1],weight_low[2],weight_low[3])
```
